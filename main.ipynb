```python
import pandas as pd 
df = pd.read_csv('uae_properties.csv')

print(df.head())
```

```python
print(df.dtypes)
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['addedOn'] = pd.to_datetime(df['addedOn'])
print(df.head(5))
```

```python
# Histogram of price
df['price'].hist()
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Boxplot of price
sns.boxplot(df['price'])
plt.title('Boxplot of Price')
plt.show()

# Histogram of bedrooms
df['bedrooms'].hist()
plt.title('Distribution of Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of property type
df['propertyType'].value_counts().plot(kind='bar')
plt.title('Frequency of Property Type')
plt.xlabel('Property Type')
plt.ylabel('Frequency')
plt.show()

# Bar plot of type
df['type'].value_counts().plot(kind='bar')
plt.title('Frequency of Type')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of price vs bedrooms
plt.scatter(df['bedrooms'], df['price'])
plt.title('Scatter Plot of Bedrooms vs Price')
plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.show()


```

```python
# Boxplot of price by property type
sns.boxplot(x='propertyType', y='price', data=df)
plt.title('Boxplot of Price by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Price')
plt.show()

# Boxplot of price by type
sns.boxplot(x='type', y='price', data=df)
plt.title('Boxplot of Price by Type')
plt.xlabel('Type')
plt.ylabel('Price')
plt.show()
```

```python
# Pairplot
sns.pairplot(df)
plt.title('Pairplot of Dataframe')
plt.show()
```

```python
# Boxplot to detect outliers in price
sns.boxplot(df['price'])
plt.title('Boxplot to Detect Outliers in Price')
plt.show()

# Z-score method
from scipy.stats import zscore
df['z_score'] = zscore(df['price'])
outliers = df[df['z_score'].abs() > 3]
print(outliers)
```
